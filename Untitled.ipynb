```python
from sagemaker.huggingface.model import HuggingFaceModel
from sagemaker.serverless import ServerlessInferenceConfig
```

```python
s3_location = "s3://bert-base-portuguese-cased/model.tar.gz"
```

```python
from sagemaker.huggingface.model import HuggingFaceModel

iam_role = "arn:aws:iam::506568066138:role/bert-portuguese-cased"

sess = sagemaker.Session()

# create Hugging Face Model Class
huggingface_model = HuggingFaceModel(
   #env=hub,
   model_data=s3_location,       # path to your model and script
   role=iam_role,                # iam role with permissions to create an Endpoint
   transformers_version="4.12",  # transformers version used
   pytorch_version="1.9",        # pytorch version used
   py_version='py38',            # python version used
   sagemaker_session=sess,
)

serverless_config = ServerlessInferenceConfig(
    memory_size_in_mb=3072, max_concurrency=1,
)

# deploy the endpoint endpoint
predictor = huggingface_model.deploy(
    serverless_inference_config=serverless_config
#     initial_instance_count=1,
#     instance_type="ml.g4dn.xlarge"
    )
```

```python
data = {
    "inputs": [
        "Eai rapaziada, tudo de boa? Eu to feliz hoje",
        "Olá amigos, como vocês estão? bão demais, fala ai",
        "Bom dia a todos! Espero que tenham um ótimo dia.",
        "Boa tarde, pessoal! Alguém tem alguma novidade para compartilhar?",
    ] * 25
}


print(len(data["inputs"]))
```

```python

import time

start_time = time.time()
res = predictor.predict(data=data)
print(time.time()  - start_time)
print(res)
```
